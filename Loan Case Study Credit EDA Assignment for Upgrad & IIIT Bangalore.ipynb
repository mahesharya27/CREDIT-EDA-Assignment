# Univariate Analysis

```python
# Plotting to TARGET variable to understand Defaulter and non-Defaulter count.
plt.figure(figsize=[4,5])
sns.set_theme(style="darkgrid")
sns.countplot(inp1_appdata.TARGET,palette=["green","red"])
plt.title("Frequency of TARGET Variable")
plt.xlabel("Non-Defaulters        Defaulters")
plt.ylabel("count")
plt.show()
# By plotting the countplot shows that Non-defaulters are more as compared to Defaulters.
```

```python
# Checking if
plt.figure(figsize=[4,5])
sns.set_theme(style="whitegrid")
sns.countplot(inp1_appdata.CODE_GENDER,palette=["darkblue","darkgreen"])
plt.title("Frequency of Gender Variable")
plt.xlabel("Male                     Female")
plt.ylabel("count")
plt.show()
# By plotting the countplot shows that Female are more as compared to Male.
```

```python
# Checking the ORGANIZATION_TYPE Column whether it has affect on Defaulter or Non-Defaulter.
inp1_appdata["ORGANIZATION_TYPE"].value_counts()
```

```python
# Since 55374 rows has "XNA" Which has no Meaning it can be Removed.
inp1_appdata=inp1_appdata.drop(inp1_appdata.loc[inp1_appdata["ORGANIZATION_TYPE"]=="XNA"].index)
inp1_appdata["ORGANIZATION_TYPE"].value_counts()
```

```python
# Count Plot to Observe the Income Type of Non-Defaulters.
sns.set(style="darkgrid")
sns.set(rc={"figure.figsize":(12,6)})

plt.title("Distribution of Income Type for Non-Defaulter")
plt.xticks(rotation=90)
plt.yscale("log")
ax=sns.countplot(data=cust_nondefaulter,x="NAME_INCOME_TYPE",hue="CODE_GENDER",order=cust_nondefaulter["NAME_INCOME_TYPE"].value_counts().index,palette="viridis")
```

```python
# Count Plot to Observe the Income Type of Defaulters.
sns.set(style="darkgrid")
sns.set(rc={"figure.figsize":(12,6)})

plt.title("Distribution of Income Type for Defaulter")
plt.xticks(rotation=90)
#plt.ylim(0,100000)
plt.yscale("log")
ax=sns.countplot(data=cust_defaulter,x="NAME_INCOME_TYPE",hue="CODE_GENDER",order=cust_nondefaulter["NAME_INCOME_TYPE"].value_counts().index,palette="hls")
```

```python
# Checking the Education-Level of Non-Defaulters and Defaulters.
fig, (p1,p2) = plt.subplots(nrows=2, ncols=1)

plt.figure(figsize=(12,12))

p1 = sns.countplot(ax=p1, data = cust_nondefaulter, x="ORGANIZATION_TYPE",order=cust_nondefaulter["ORGANIZATION_TYPE"].value_counts().index,palette="terrain")
p1.set_xlabel("Non Defaulters")
p1.set_xticklabels(p1.get_xticklabels(),rotation=90)

p2 = sns.countplot(ax=p2, data = cust_defaulter, x="ORGANIZATION_TYPE",order=cust_defaulter["ORGANIZATION_TYPE"].value_counts().index,palette="copper")
p2.set_xlabel("Defaulters")
p2.set_xticklabels(p2.get_xticklabels(),rotation=90)

plt.show()
```

```python
# Checking the Education level of Non-Defaulters and Defaulters
fig, (p1,p2) = plt.subplots(nrows=1, ncols=2)

plt.figure(figsize=(12,12))

p1 = sns.countplot(ax=p1, data = cust_nondefaulter, x="NAME_EDUCATION_TYPE",palette="crest")
p1.set_xlabel("Non Defaulters")
p1.set_xticklabels(p1.get_xticklabels(),rotation=90)

p2 = sns.countplot(ax=p2, data = cust_defaulter, x="NAME_EDUCATION_TYPE",palette="flare")
p2.set_xlabel("Defaulters")
p2.set_xticklabels(p2.get_xticklabels(),rotation=90)

plt.show()
# subplots shows that "Secondary/Secondary special" Education customers are maximum Non-Defaulters and Maximum Defaulters too.
```

```python
# Checking the Annual Income Column spread for defaulters 

plt.figure(figsize=(12,6))
sns.set(style="dark")

plt.subplot(1,2,1)
plt.title("Non Defaulters")
plt.xlabel("Income of Non Defaulters ")
plt.ylabel("Loan Count")
s=sns.boxplot(y=cust_nondefaulter["AMT_INCOME_TOTAL"])

plt.subplot(1,2,2)
plt.title("Defaulters")
plt.xlabel("Income of Defaulters")
plt.ylabel("Loan Count")
s=sns.boxplot(y=cust_defaulter["AMT_INCOME_TOTAL"])

plt.show()
```

```python
# Checking  Non-Defaulter and Defaulter have how many Childrens.
fig, ax=plt.subplots(figsize=[12,6])
ax.scatter(cust_nondefaulter["CNT_CHILDREN"],cust_nondefaulter["AMT_CREDIT"])
ax.set_xlabel("No. of Children")
ax.set_ylabel("Loan Applied by Non Defaulters")

fig, ax=plt.subplots(figsize=[12,6])
ax.scatter(cust_defaulter["CNT_CHILDREN"],cust_defaulter["AMT_CREDIT"])
ax.set_xlabel("No. of Children")
ax.set_ylabel("Loan Applied by Defaulters")

plt.show()
```

```python
# Checking which Occupation Type tend to default the Loan.
fig, (p1,p2) = plt.subplots(nrows=1, ncols=2)

p1 = sns.countplot(ax=p1, data = cust_nondefaulter, x="OCCUPATION_TYPE",palette="terrain")
p1.set_xlabel("Non Defaulters")
p1.set_xticklabels(p1.get_xticklabels(),rotation=90)

p2 = sns.countplot(ax=p2, data = cust_defaulter, x="OCCUPATION_TYPE",palette="cubehelix")
p2.set_xlabel("Defaulters")
p2.set_xticklabels(p2.get_xticklabels(),rotation=90)

plt.show()
# Count plots below shaws that Laborers are most tends to default the loan and after them Sales staff tend to default the Loan.
```

```python
# Checking the Family status of Defaulters and Non-Defaulters.
fig, (p1,p2) = plt.subplots(nrows=1, ncols=2)

p1 = sns.countplot(ax=p1, data = cust_nondefaulter, x="NAME_FAMILY_STATUS",palette="crest")
p1.set_xlabel("Non Defaulters")
p1.set_xticklabels(p1.get_xticklabels(),rotation=90)

p2 = sns.countplot(ax=p2, data = cust_defaulter, x="NAME_FAMILY_STATUS",palette="flare")
p2.set_xlabel("Defaulters")
p2.set_xticklabels(p2.get_xticklabels(),rotation=90)

plt.show()
#Subplots below shows that Married Customers are most in Non-Defaulters and Defaulters too.
```

```python
# Select the Columns for Defaulter Correlation
corr_cols=["CNT_FAM_MEMBERS","AMT_INCOME_TOTAL","AMT_CREDIT","CNT_CHILDREN"]
co_defaulters=cust_defaulter[corr_cols]
corr_defaulters=co_defaulters.corr()
round(corr_defaulters,5)
```

```python
# Plotting Heat Map for more understanding.
plt.figure(figsize=[6,6])
sns.heatmap(corr_defaulters,cmap="RdYlGn",annot=True)
plt.show()
```

```python
# Analysing How many Previous loans Approved/Refused/Canceled/Unused Offer.
plt.figure(figsize=[6,7])
sns.set_theme(style="whitegrid")

sns.countplot(inp2_prevdata.NAME_CONTRACT_STATUS,palette=["blue","green"])
plt.title("Status of Previous Application")
plt.xlabel("Contract Status")
plt.ylabel("Number of Contracts")
plt.show()
```

```python
# Checking for What was the most Approved/Refused/Canceled/Unused Offer Loan Type.

loan_approved=inp2_prevdata[inp2_prevdata.NAME_CONTRACT_STATUS=="Approved"]
loan_refused=inp2_prevdata[inp2_prevdata.NAME_CONTRACT_STATUS=="Refused"]
loan_cancelled=inp2_prevdata[inp2_prevdata.NAME_CONTRACT_STATUS=="Cancelled"]
loan_unused=inp2_prevdata[inp2_prevdata.NAME_CONTRACT_STATUS=="Unused Offer"]
```

```python
# Plotting Counplot for Approved/Refused/Canceled/Unused Offer Loan Type.
fig, (p1,p2) = plt.subplots(nrows=1, ncols=2)
plt.figure(figsize=[14,6])

p1 = sns.countplot(ax=p1, data = loan_approved, x="NAME_PORTFOLIO",palette="crest")
p1.set_xlabel("Approved")
p1.set_xticklabels(p1.get_xticklabels(),rotation=90)

p2 = sns.countplot(ax=p2, data = loan_refused, x="NAME_PORTFOLIO",palette="flare")
p2.set_xlabel("Refused")
p2.set_xticklabels(p2.get_xticklabels(),rotation=90)

plt.show()
# From the below Count plot shows that the Maximum Approved loan is POS type and Maximum Refused loan is of Cash Loan Type
```

```python
# Analysis of Channel Type

fig, (p1,p2) = plt.subplots(nrows=1, ncols=2)

p1 = sns.stripplot(ax=p1, x= loan_approved["CHANNEL_TYPE"], y=loan_approved["AMT_CREDIT"],palette="crest")
p1.set_xlabel("Approved Customers Channel")
p1.set_xticklabels(p1.get_xticklabels(),rotation=90)

p2 = sns.stripplot(ax=p2, x= loan_refused["CHANNEL_TYPE"], y=loan_refused["AMT_CREDIT"],palette="flare")
p2.set_xlabel("Refused Customers Channel")
p2.set_xticklabels(p2.get_xticklabels(),rotation=90)

plt.show()
```

```python
# Finding Correlation for Defaulters.

corr_cols1=["CNT_CHILDREN","AMT_INCOME_TOTAL","AMT_CREDIT","AMT_ANNUITY","REGION_POPULATION_RELATIVE","DAYS_BIRTH","LIVE_REGION_NOT_WORK_REGION",
            "REG_CITY_NOT_LIVE_CITY","REG_CITY_NOT_WORK_CITY","LIVE_CITY_NOT_WORK_CITY"]
co_defaulters1=cust_defaulter[corr_cols1]
corr_defaulters1=co_defaulters1.corr()
round(corr_defaulters1,2)
```

```python
# For more Clarification Plotting Heatmap for finding Correlation for Defaulters.
plt.figure(figsize=[16,6])
sns.heatmap(corr_defaulters1,cmap="RdYlGn",annot=True)
plt.show()
```

```python
# Finding Correlation for Non-Defaulters.
corr_cols2=["CNT_CHILDREN","AMT_INCOME_TOTAL","AMT_CREDIT","AMT_ANNUITY","REGION_POPULATION_RELATIVE","DAYS_BIRTH","LIVE_REGION_NOT_WORK_REGION",
            "REG_CITY_NOT_LIVE_CITY","REG_CITY_NOT_WORK_CITY","LIVE_CITY_NOT_WORK_CITY"]
co_nondefaulters=cust_nondefaulter[corr_cols2]
corr_nondefaulters=co_nondefaulters.corr()
round(corr_nondefaulters,2)
```

```python
# For more Clarification Plotting Heatmap for finding Correlation for Non-Defaulters.
plt.figure(figsize=[16,6])
sns.heatmap(corr_nondefaulters,cmap="RdYlGn",annot=True)
plt.show()
```
